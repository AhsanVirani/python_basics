```python
df['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
def func1(ser):
    return (ser.str.contains('Python').sum())
users_python = country_grp['LanguageWorkedWith'].apply(func1)
```

```python
total_users = df.groupby('Country')['Country'].count()
```

```python
per_py_users = (country_grp['LanguageWorkedWith'].apply(func1)/total_users)*100
```

```python
per_py_users['United States']
```

```python
# Another method to do the same
```

```python
tot_users = df['Country'].value_counts()
```

```python
new_data = pd.concat([tot_users, users_python], axis = 1)
```

```python
 new_data.head(50)
```

```python
 new_data['perc'] = new_data['LanguageWorkedWith']/new_data['Country']*100
```

```python
new_data.rename(columns={"Country":"Total Users", "LanguageWorkedWith":"Python Users", "perc":"Perc Pyth Users"}, inplace = True)
```

```python
new_data.sort_values(by="Perc Pyth Users", ascending = False, inplace = True)
```
